```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from venn import venn

import sys
sys.path.append('../script')
from GO_Kcat_analysis import *
```

# Input and output Data

iML1515R   

```python
#basic path
base_csv_path = "../data/iML1515R/" #5883 reactions, 5338 enzyme-related   

kcat_csv_files = {
    "DLKcat": "reaction_kcat_MW_DLKcat.csv",
    "TurNup": "reaction_kcat_MW_TurNup.csv",
    "UniKP": "reaction_kcat_MW_UniKP.csv"
}

# reaction-kcat data
go_hkp_kcat = load_json('../analysis/DeepGO-SE/iML1515R/go_kcat_mean_parent_process_Total_median.json')
dlkcat_kcat = load_csv(base_csv_path + kcat_csv_files["DLKcat"], "reactions")
turnup_kcat = load_csv(base_csv_path + kcat_csv_files["TurNup"], "reactions")
unikp_kcat = load_csv(base_csv_path + kcat_csv_files["UniKP"], "reactions")

# output path
venn_output_path = "../analysis/DeepGO-SE/iML1515R/data_comparison_venn_diagram.png"
```

# Venn diagram drawing

```python
data = {
    'GO-HKP': set(go_hkp_kcat.keys()),
    'DLKcat': set(dlkcat_kcat.index),
    'TurNup': set(turnup_kcat.index),
    'UniKP': set(unikp_kcat.index)
}

plt.rcParams.update({
    'figure.facecolor': 'white',
    'axes.facecolor': 'white',
    'text.color': 'black',
    'axes.labelcolor': 'black',
    'xtick.color': 'black',
    'ytick.color': 'black',
    'legend.facecolor': 'white',
    'legend.edgecolor': 'black'
})

plt.figure(figsize=(10, 8))
venn_diagram = venn(data, fmt="{size:.0f}", cmap="Pastel1", fontsize=12, legend_loc="best")
plt.title("Venn Diagram of Reaction IDs from Different Sources", fontsize=14)

for text in plt.gcf().findobj(lambda x: hasattr(x, 'get_color') and x.get_color() == 'white'):
    text.set_color('black')

plt.savefig(venn_output_path, dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
